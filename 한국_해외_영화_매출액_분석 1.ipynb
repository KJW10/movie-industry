### (1) 2019년 지역별

```python
DF_19K = df_2019.groupby('시도명')['한국영화매출금액'].sum()
DF_19F = df_2019.groupby('시도명')['해외영화매출금액'].sum()
```

```python
DF_19K = df_2019.groupby('시도명')['한국영화관람객수'].sum()
DF_19F = df_2019.groupby('시도명')['해외영화관람객수'].sum()
DF_19K = DF_19K.to_frame()
DF_19F = DF_19F.to_frame()
DF_19K['해외영화관람객수'] = DF_19F['해외영화관람객수']
DF_19 = DF_19K
```

```python
A = pd.DataFrame({'시도명':DF_19.index,
                  '한국영화관람객수':DF_19['한국영화관람객수'].values,
                  '해외영화관람객수':DF_19['해외영화관람객수'].values})

A[['한국영화관람객수', '해외영화관람객수']] = round(A[['한국영화관람객수', '해외영화관람객수']]/1000000,1)
A
```

```python
plt.figure(figsize = (15, 7))

plt.bar(A.index, A['한국영화관람객수'],
        width = 0.3, label = '한국영화관람객수(단위:백 만)')
plt.bar(A.index + 0.3, A['해외영화관람객수'],
         width = 0.3, label = '해외영화관람객수(단위:백 만)')
plt.title('2019년 한국/해외영화 지역별 관객수', fontsize = 20)
plt.xlabel('시도명', fontsize = 17)
plt.ylabel('관객수(단위:백 만)', fontsize = 17)
plt.xticks(A.index,A['시도명'].unique(),rotation = 25)
plt.legend()
plt.show()
```
