```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'NanumGothic'
```

# *극장수/스크린수/좌석수

```python
df = pd.read_excel('극장스크린.xlsx')
df1 = df.copy()
df1.index = df1['연도']
df_th = df1['극장수']
df_sc = df1['스크린수']
df_se = df1['좌석수']
```

```python

fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (21, 5))

ax[0].plot(df_th,
           marker = 'o')
ax[1].plot(df_sc,
           marker = 'o')
ax[2].plot(df_se,
           marker = 'o')

ax[0].set_title('연도별 극장수')
ax[1].set_title('연도별 스크린수')
ax[2].set_title('연도별 좌석수')

ax[0].set_xlabel('연도')
ax[1].set_xlabel('연도')
ax[2].set_xlabel('연도')

ax[0].set_ylabel('극장수(단위: 개)')
ax[1].set_ylabel('스크린수(단위: 개)')
ax[2].set_ylabel('좌석수(단위: 개)')

plt.show()
```

* 극장 수/스크린 수/좌석 수에 대한 분석을 실시해본 결과, 극장 수와 스크린 수는
2020년을 기점으로 계속해서 상승하고 있는 반면, 좌석수는 2020-2021년에 증가하였다가 점차 감소하는 추세를 보이고 있었음
* 이에 대한 자료조사를 해본 결과 현재 영화관 산업은 팬데믹 시기를 기점으로 현재 좋지 못한 상황에 대응하기 위해 극장 관람 환경 및 시스템의 변화를 도모
* 관람 경험을 차별화 및 고급화하는 전략이 원인으로써 작용하여 극장 수/스크린 수와 대조적으로 좌석 수가 감소하고 있는 것을 알 수 있었습니다.
