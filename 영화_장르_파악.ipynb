```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
DF = pd.read_excel('장르2018.xlsx')
DF
```

```python
sns.set_theme(context='notebook',   # 매체: paper, talk, poster     #sns seaborn의 설정
              style='white',     # 기본 내장 테마
              palette='pastel',       # 그래프 색
              font='NanumGothic', # 글꼴 종류
              font_scale=0.8)
```

```python
DF.columns = ['순위', '영화명', '개봉일', '매출액', '매출액', '누적매출액', '관객수', '누적관객수', '스크린수',
       '상영횟수', '대표국적', '국적', '제작사', '배급사', '등급', '장르', '감독', '배우']
DF_1 = DF[DF['장르'].isna() == False]
DF_1 = DF_1[DF_1['개봉일'].dt.year >= 2017]
DF_2019 = DF_1[DF_1['개봉일'].dt.year == 2019]
DF_2018= DF_1[DF_1['개봉일'].dt.year == 2018]
```

# 2018년

```python
DF_2018['주장르'] = DF_2018['장르'].str.split(',').str[0]
DF_2018 = DF_2018.groupby('주장르')['매출액'].sum()
DF_2018['매출액'].sort_values(ascending = False).head(10)
```

```python
plt.figure(figsize = (10, 10))
plt.pie(DF_2018['매출액'].sort_values(ascending = False).head(5),
        labels = DF_2018['매출액'].sort_values(ascending = False).head(5).index,
        autopct = '%.1f%%',
        startangle = 90,
        textprops = {'fontsize':20})
plt.title('2018년 장르별 매출액 비율(상위 5개)',fontsize = 20)
plt.show()
```

```python
DF_2019['주장르'] = DF_2019['장르'].str.split(',').str[0]
DF_2019 = DF_2019.groupby('주장르')['매출액'].sum()
DF_2019['매출액'].sort_values(ascending = False).head(10)
```
